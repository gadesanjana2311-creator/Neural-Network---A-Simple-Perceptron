**Question 2:** Explain the basic architecture and functioning of a Perceptron. What are its
limitations?

**Answer:**
Perceptron: Basic Architecture and Functioning

The Perceptron is one of the earliest and simplest types of artificial neural networks. It was introduced in 1958 by Frank Rosenblatt. It is mainly used for binary classification problems.

***1. Basic Architecture of a Perceptron***


A Perceptron consists of the following components:

1. Input Layer (x₁, x₂, ..., xₙ)

Receives input features.

2. Weights (w₁, w₂, ..., wₙ)

Each input has an associated weight representing its importance.

3. Bias (b)

A constant added to adjust the output independently of inputs.

4. Summation Function

Computes the weighted sum of inputs:

𝑧=𝑤1𝑥1+𝑤2𝑥2+...+𝑤𝑛𝑥n

5. Activation Function (Step Function)

Produces the final output:

If z≥0, output = 1

If z<0, output = 0

***2. Working (Functioning) of a Perceptron***

The perceptron works in the following steps:

1. Receive Inputs

Input values are fed into the model.

2. Multiply by Weights

Each input is multiplied by its corresponding weight.

3. Compute Weighted Sum

All weighted inputs are summed together with bias.

4. Apply Activation Function

The step function determines whether the neuron "fires" (outputs 1) or not (outputs 0).

5. Learning (Weight Update Rule)

***3. Limitations of the Perceptron***

Despite its importance, the perceptron has several limitations:

1. Only Solves Linearly Separable Problems

It can classify data only if a single straight line (or hyperplane) can separate the classes.

It cannot solve non-linear problems such as the XOR problem.

2. Binary Classification Only

The original perceptron handles only two classes (0 or 1).

3. No Hidden Layers

It is a single-layer neural network, so it cannot learn complex patterns.

4. Uses Step Activation Function

The step function is not differentiable, making it unsuitable for advanced gradient-based optimization used in deep learning.

**Question 3:** Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

**Answer:**

Purpose of Activation Function in Neural Networks

An activation function determines whether a neuron should be activated (i.e., produce an output) or not. It introduces non-linearity into the network, allowing neural networks to learn and model complex patterns.

Why Activation Functions Are Important

1. Introduce Non-Linearity

Without activation functions, a neural network would behave like a linear model, no matter how many layers it has.

2. Control Output Range

Activation functions scale outputs into specific ranges (e.g., 0–1 or −1–1).

3. Enable Learning of Complex Patterns

They allow deep networks to approximate highly non-linear functions.

4. Support Backpropagation

Differentiable activation functions allow gradient-based optimization.

| ------------------ | --------------------- | ------------- | --------------------------- |

| Vanishing Gradient | Yes                   | Yes           | Reduced                     |
| Computational Cost | Moderate              | Moderate      | Low                         |
| Common Usage       | Output layer (binary) | Hidden layers | Hidden layers (most common) |


**Question 4:** What is the difference between Loss function and Cost function in neural
networks? Provide examples.

**Answer:**

Key Differences of Loss function and Cost function

| Aspect     | Loss Function                 | Cost Function                              |
| ---------- | ----------------------------- | ------------------------------------------ |
| Definition | Error for a single data point | Average/total error over entire dataset    |
| Scope      | Individual sample             | Entire dataset                             |
| Usage      | Used during weight updates    | Used to evaluate overall model performance |
| Example    | Squared error for one sample  | Mean Squared Error over all samples        |

**Simple Example**

Suppose we have 3 data points and squared errors:

Sample 1 loss = 2

Sample 2 loss = 4

Sample 3 loss = 6

Loss:

Each individual value (2, 4, 6) is a loss.

Cost:

Cost=(2+4+6)/3=4


So, 4 is the cost, while 2, 4, and 6 are individual losses.

Role of Optimizers in Neural Networks

An optimizer is an algorithm used to minimize the cost (loss) function by updating the model’s weights during training.

Why Optimizers Are Important

1. Adjust weights to reduce prediction error

2. Speed up convergence

3. Improve model accuracy

4. Help escape poor local minima (in advanced methods)

```python
#Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.

import numpy as np

# Step 1: Define the AND gate dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])  # AND outputs

# Step 2: Initialize parameters
np.random.seed(42)
weights = np.random.randn(2)
bias = np.random.randn()
learning_rate = 0.1
epochs = 10

# Step 3: Define activation function (Step function)
def step_function(z):
    return 1 if z >= 0 else 0

# Step 4: Training the perceptron
for epoch in range(epochs):
    print(f"\nEpoch {epoch + 1}")
    for i in range(len(X)):
        # Linear combination
        z = np.dot(X[i], weights) + bias

        # Prediction
        y_pred = step_function(z)

        # Error
        error = y[i] - y_pred

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}")

# Step 5: Testing the trained perceptron
print("\nFinal Weights:", weights)
print("Final Bias:", bias)

print("\nTesting AND Gate:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    y_pred = step_function(z)
    print(f"Input: {X[i]} -> Output: {y_pred}")
```

```python
#Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 400)

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Create subplots
plt.figure(figsize=(12, 8))

# Sigmoid Plot
plt.subplot(3, 1, 1)
plt.plot(x, y_sigmoid, color='blue')
plt.title("Sigmoid Activation Function")
plt.grid(True)

# ReLU Plot
plt.subplot(3, 1, 2)
plt.plot(x, y_relu, color='green')
plt.title("ReLU Activation Function")
plt.grid(True)

# Tanh Plot
plt.subplot(3, 1, 3)
plt.plot(x, y_tanh, color='red')
plt.title("Tanh Activation Function")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# 1. Load MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# 2. Preprocess the data
x_train = x_train.reshape(-1, 28 * 28).astype("float32") / 255
x_test = x_test.reshape(-1, 28 * 28).astype("float32") / 255

# 3. Build the model
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 4. Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 5. Train the model
history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=32,
    verbose=1
)

# 6. Print training accuracy
train_loss, train_accuracy = model.evaluate(x_train, y_train, verbose=0)
print(f"\nTraining Accuracy: {train_accuracy:.4f}")
```

```python
#Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# 1. Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# 2. Preprocess the data
x_train = x_train.reshape(-1, 28 * 28).astype("float32") / 255
x_test = x_test.reshape(-1, 28 * 28).astype("float32") / 255

# 3. Build the model
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 4. Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 5. Train the model
history = model.fit(
    x_train, y_train,
    epochs=15,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# 6. Plot Loss and Accuracy Curves
plt.figure(figsize=(12, 5))

# Loss Curve
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# Accuracy Curve
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
